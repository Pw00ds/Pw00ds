# Step 5: Visualize & Interpret

## Scatter Plot
By creating a scatter plot we can visualize the relationship between the social support and happiness. 

I hypothesize that these two variables will have a positive relationship - meaning as the `Explained by: Social support` variable increases, the `Happiness score` variable should also increase. 

We can do this using the `plt.scatter()` function from the matplotlib library.

```python
#Plot a scatter plot (based on Data Camp tutorial)
plt.scatter(df["Explained by: Social support"], df["Happiness score"])

#Label Axes


#Show
plt.show()
```

We can also use the `.axes()` object to plot our data and customize our plot. Some resources you may want to look at on your free time are found below.

Resources:
*   [What is `fig, ax = plt.subplots()`?](https://stackoverflow.com/questions/34162443/why-do-many-examples-use-fig-ax-plt-subplots-in-matplotlib-pyplot-python)
*   [Great Introduction to `.axes()` objects](https://www.earthdatascience.org/courses/scientists-guide-to-plotting-data-in-python/plot-with-matplotlib/introduction-to-matplotlib-plots/customize-plot-colors-labels-matplotlib/)
*   [Matplotlib Reference for Axes Object Commands](https://matplotlib.org/stable/api/axes_api.html)

```python
# Define plot space for figure and axes objects
fig, ax = plt.subplots(figsize=(10, 6))

#Create a scatter plot visualize the relationship between Days in Top 10 and Viewership Score
ax.scatter(x = df['Explained by: Social support'], 
           y = df['Happiness score'], 
           color = "dodgerblue",  #change color
           s= df['RANK'],         #change size based on a variable
           alpha = 0.5,           #transparency of dots
           marker = 'p') #more  markers here https://matplotlib.org/3.3.3/api/markers_api.html

#labels and settings
ax.set_title("Happiness vs Social Support", fontsize = 20) #title
ax.set_xlabel("Social Support", fontsize=20) # x labels
ax.set_ylabel("Happiness", fontsize=20) #y labels
plt.show() #show
```

**Interpretation:** From the above scatterplot we can see that our hypothesis was correct. The variables have a positive relationship as shown in as "Social Support" increases, "Happiness" also increases.
